# Web Scraping from OLX

## Scraping Mercedes Benz Car Price and Specification

```python
# Library for scraping process

import numpy as np
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# scrapping script

Title = []
Price = []
Linked = []
Merek = []
Varian = []
Model = []
Tahun = []
Jarak_tempuh = []
Tipe_bbm = []
Warna =[]
Transmisi = []
Kapasitas_mesin = []
fixed_parameters = ['Merek','Varian', 'Model', 'Tahun', 'Jarak tempuh', 'Tipe bahan bakar', 'Warna', 'Transmisi', 'Kapasitas mesin']

pages = 1
while pages < 100:
    try :
        html = urlopen("https://www.olx.co.id/mobil-bekas_c198/q-mercedes-benz?page=%s" %(pages)).read()
        soup = BeautifulSoup(html, "lxml")
        produk = soup.find_all("li","EIR5N")
        for p in produk:
            if (p.find("span","_89yzn") != None) :
                title_temp = p.find("span","_2tW1I").get_text()
                price_temp = p.find("span","_89yzn").get_text()
                link = "https://www.olx.co.id"+p.find("a",href=True)['href']

                html_item = urlopen(link).read()
                soup_item = BeautifulSoup(html_item, "lxml")
                parameters_item = soup_item.find_all("span","_25oXN")
                produk_item = soup_item.find_all("span","_2vNpt")

                parameters = []  
                for param in parameters_item:
                    parameters.append(param.get_text())

                list_item = []
                for item in produk_item:
                    list_item.append(item.get_text())


                new_parameters = [item for item in parameters  if item in fixed_parameters ]

                if len(new_parameters) == len(fixed_parameters):
                    new_list_item = []
                    for ele in new_parameters:
                        new_list_item.append(list_item[parameters.index(ele)])

                    sorted_parameters = sorted(new_parameters, key=lambda x: fixed_parameters.index(x))

                    sorted_list_item = []
                    for ele in sorted_parameters:
                        sorted_list_item.append(new_list_item[new_parameters.index(ele)])

                    Title.append(title_temp)
                    Price.append(price_temp)
                    Linked.append(link)
                    Merek.append(sorted_list_item[0])
                    Varian.append(sorted_list_item[1])
                    Model.append(sorted_list_item[2])
                    Tahun.append(sorted_list_item[3])
                    Jarak_tempuh.append(sorted_list_item[4])
                    Tipe_bbm.append(sorted_list_item[5])
                    Warna.append(sorted_list_item[6])
                    Transmisi.append(sorted_list_item[7])
                    Kapasitas_mesin.append(sorted_list_item[8])

#                     print(sorted_parameters)
#                     print(sorted_list_item)

                    new_list_item.clear()
                    sorted_list_item.clear()

                else:
                    pass

                parameters.clear()
                list_item.clear()

            else:
                pass
    except:
        pass
    
    print ("check loop, page %s done" %(pages))
    pages += 1

    
data_olx = {
    "Title" : Title,
    "Link" : Linked,
    "Price" : Price,
    "Merek" : Merek,
    "Varian" : Varian,
    "Model" : Model,
    "Tahun" : Tahun,
    "Jarak Tempuh" : Jarak_tempuh,
    "Tipe BBM" : Tipe_bbm,
    "Warna" : Warna,
    "Transmisi" : Transmisi,
    "Kapasitas Mesin" : Kapasitas_mesin
}
```

```python
df = pd.DataFrame(data_olx)
df
```

```python
len(df['Link'].unique())
```
